```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df= pd.read_excel('/content/watson_healthcare_modified.xlsx')
df.head()

```

```python
# Bar graph
plt.figure(figsize=(8, 6))
plt.bar(df['JobRole'], df['MonthlyIncome'], color='darkblue')
plt.xlabel('JobRole')
plt.ylabel('MonthlyIncome')
plt.title('Employees Salary')
plt.show()

```

```python
# Pie chart
plt.figure(figsize=(8, 8))
department= df['JobRole'].value_counts()
plt.pie(department, autopct='%1.1f%%', startangle=90, colors=['purple', 'yellow', 'teal', 'red'])
plt.title('Employee Distribution')
plt.show()
```

```python
count_df = df.groupby(['EducationField', 'JobRole']).size().reset_index(name='Count')
plt.figure(figsize=(10, 6))
count_df.plot(x='JobRole', y='Count', kind='line', color='skyblue')
plt.xlabel('Job Roles')
plt.ylabel('Count')
plt.title('Count of Employees by Education Field and Job Role')
plt.xticks(rotation=45)
plt.show()
```
